```python
piece_size = 1000000

line_count = 0

for piece in pd.read_csv("./searches.csv", chunksize = piece_size, sep = '^'):
    line_count = line_count + len(piece.index)
    
print('searches.csv has '+ str(line_count)+' lines')
```
