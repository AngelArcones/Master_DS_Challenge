```python
import pandas as pd
bi = pd.read_csv('bookings.csv.bz2', sep='^', usecols=columns_of_interest, nrows=9999, chunksize=3000)

all_chunks = pd.DataFrame()

for i,b in enumerate(bi):
    print(i, len(b))
    b = b.dropna()
    b = b[ b['year']==2013]
    del b['year']
    result = b.groupby(['arr_port']).sum()
    all_chunks = all_chunks.append(result)

all_chunks.head(10)

result_all = all_chunks.reset_index().groupby('arr_port').sum().sort_values(by='pax', ascending=False).reset_index()
result_all.head(10)
```

```python
%%time
import pandas as pd
bi = pd.read_csv('bookings.csv.bz2', sep='^', usecols=columns_of_interest, chunksize=1000000)

all_chunks = pd.DataFrame()

for i,b in enumerate(bi):
    print(i, len(b))
    b = b.dropna()
    b = b[ b['year']==2013]
    del b['year']
    result = b.groupby(['arr_port']).sum()
    all_chunks = all_chunks.append(result)

all_chunks.head(10)

result_all = all_chunks.reset_index().groupby('arr_port').sum().sort_values(by='pax', ascending=False).reset_index()
result_all.head(10)
```
